```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('credit_data.csv')
df.head()
```

## Imbalanced 0 = 284315, 1 = 492 not balanced in our data
### 0 ---> Legit Transactions
### 1 ---> Fraudulent Transactions

```python
df['Class'].value_counts()
```

```python
# separating the legit and fraudulent transactions
legit = df[df['Class']==0]
fraudulent = df[df['Class']==1]
```

```python
print(legit.shape)
print(fraudulent.shape)
```

Undersampling

build a sample dataset containing similar distribution of Legit and Fraudulent

Number of Fraudulent Transactions ---> 492 (radomly)

```python
legit_sample = legit.sample(n=492)
legit_sample
```

Concattenate the two DataFrame

```python
new_dataset = pd.concat([legit_sample, fraudulent],axis=0)
new_dataset.shape
```
